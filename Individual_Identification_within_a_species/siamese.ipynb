# Class Imbalance

```python
# Print percentage of examples in each class
def percentage_distribution(df : pd.DataFrame, column : str) -> pd.Series:
    series = df[column].value_counts()
    series = (series / series.sum()) * 100
    
    # Append count column to the series
    series = pd.concat([series, df[column].value_counts()], axis=1)
    series.columns = ['percentage', 'count']
    return series

# Plot a histogram of the distribution of examples in each class
def plot_distribution(df : pd.DataFrame, column : str):
    series = percentage_distribution(df, column)
    series['percentage'].plot(kind='bar', figsize=(10, 5), title='Percentage distribution of examples in each class')
```

# Get Cropped Images of the detected animal using an object detection model 

```python

import cv2
import torch
from ultralytics import YOLO
import matplotlib.pyplot as plt

def predict_image(image_path, model_path):
    """
    Load a saved YOLO model and run predictions on a single image. 
    Also returns the cropped image based on detected bounding boxes.
    
    Parameters:
    - image_path (str): Path to the image for detection.
    - model_path (str): Path to the YOLO model (e.g., 'yolov8n.pt').
    
    Returns:
    - results (list): Prediction results from YOLO.
    - cropped_image (ndarray): Cropped image based on the detected bounding box.
    """
    
    # Load the YOLO model
    model = YOLO(model_path)
    
    # Run prediction on the image
    results = model.predict(image_path)
    
    # Extract the image and bounding boxes from the results
    img = cv2.imread(image_path)  # Read image using OpenCV
    
    # Assuming the first detected object is the one we want to crop (if multiple objects, modify accordingly)
    if results[0].boxes is not None:
        # Get the coordinates of the bounding box (xmin, ymin, xmax, ymax)
        bbox = results[0].boxes[0].xyxy[0].cpu().numpy()  # Get the first bounding box
        xmin, ymin, xmax, ymax = map(int, bbox)
        
        # Draw the bounding box on the image
        img_with_bbox = img.copy()
        cv2.rectangle(img_with_bbox, (xmin, ymin), (xmax, ymax), (0, 255, 0), 2)  # Draw bounding box in green
        
        # Crop the image using the bounding box
        cropped_image = img[ymin:ymax, xmin:xmax]  # Crop the image within the bounding box
        
        # Convert images from BGR to RGB (OpenCV loads images in BGR)
        img_with_bbox_rgb = cv2.cvtColor(img_with_bbox, cv2.COLOR_BGR2RGB)
        cropped_image_rgb = cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB)
        
        # Show images using matplotlib
        plt.figure(figsize=(10, 5))

        # Display the image with bounding box
        plt.subplot(1, 2, 1)
        plt.imshow(img_with_bbox_rgb)
        plt.title("Image with Bounding Box")
        plt.axis("off")
        
        # Display the cropped image
        plt.subplot(1, 2, 2)
        plt.imshow(cropped_image_rgb)
        plt.title("Cropped Image")
        plt.axis("off")
        
        plt.show()

        return results, cropped_image
    
    else:
        print("No bounding box detected.")
        return None, None

# Test the function
image_path = '/home/pinakinc/dev/umc-project/data/IPanda50/iPanda50-images/04_chengdui/04_215.jpg'
model_path = 'yolov8n.pt'
results, cropped_image = predict_image(image_path, model_path)

# Optionally, save the cropped image
if cropped_image is not None:
    output_path = 'cropped_image.jpg'
    cv2.imwrite(output_path, cropped_image)
    print(f"Cropped image saved at: {output_path}")
```

```python
import os
import cv2
import torch
from ultralytics import YOLO
import pandas as pd
from tqdm import tqdm
import numpy as np

def create_output_directories(base_path):
    """Create output directories for cropped images."""
    dirs = {
        'zebra': os.path.join(base_path, 'cropped_zebras'),
        'panda': os.path.join(base_path, 'cropped_pandas')
    }
    
    for dir_path in dirs.values():
        os.makedirs(dir_path, exist_ok=True)
        
    return dirs

def verify_path(path):
    """Verify if a path exists and is readable."""
    return os.path.exists(path) and os.path.isfile(path)

def process_image(image_path, model, output_path):
    """
    Process a single image using YOLO model and save the cropped result.
    
    Returns:
    - tuple: (bool, str) - (Success status, Error message if any)
    """
    try:
        # Verify path exists
        if not verify_path(image_path):
            return False, f"File not found: {image_path}"
            
        # Read the image
        img = cv2.imread(image_path)
        if img is None:
            return False, f"Failed to read image: {image_path}"
            
        # Run YOLO prediction
        results = model.predict(img)
        
        # Check if any detection was made
        if results[0].boxes is not None and len(results[0].boxes) > 0:
            # Get the coordinates of the bounding box with highest confidence
            boxes = results[0].boxes
            confidences = boxes.conf.cpu().numpy()
            best_box_idx = np.argmax(confidences)
            bbox = boxes[best_box_idx].xyxy[0].cpu().numpy()
            
            # Crop the image
            xmin, ymin, xmax, ymax = map(int, bbox)
            cropped_image = img[ymin:ymax, xmin:xmax]
            
            # Create output directory if it doesn't exist
            os.makedirs(os.path.dirname(output_path), exist_ok=True)
            
            # Save the cropped image
            cv2.imwrite(output_path, cropped_image)
            return True, "Success"
            
        return False, "No detections found"
        
    except Exception as e:
        return False, f"Error: {str(e)}"

def fix_panda_path(path, base_path):
    """
    Construct the correct path for panda images
    """
    # The path in the dataset is already correct relative to IPanda50 directory
    return os.path.join(base_path, 'IPanda50', path)

def process_dataset(df, dataset_type, model, output_dir, base_path=''):
    """
    Process an entire dataset and save cropped images.
    """
    successful = 0
    failed = 0
    error_log = []
    
    # Create progress bar
    progress_bar = tqdm(total=len(df), desc=f'Processing {dataset_type} images')
    
    for idx, row in df.iterrows():
        try:
            # Construct the full image path
            if dataset_type == 'zebra':
                image_path = os.path.join(base_path, row['path'])
            else:  # panda
                image_path = fix_panda_path(row['path'], base_path)
            
            # Print first few paths to debug
            if idx < 5:
                print(f"Debug - Processing path: {image_path}")
                print(f"Path exists: {os.path.exists(image_path)}")
            
            # Construct output path
            output_filename = f"{row['identity']}_{os.path.basename(row['path'])}"
            output_path = os.path.join(output_dir, output_filename)
            
            # Process the image
            success, error_msg = process_image(image_path, model, output_path)
            
            if success:
                successful += 1
            else:
                failed += 1
                error_log.append(f"Image {image_path}: {error_msg}")
                
        except Exception as e:
            failed += 1
            error_log.append(f"Exception processing {image_path}: {str(e)}")
            
        progress_bar.update(1)
    
    progress_bar.close()
    
    # Print results
    print(f"\n{dataset_type.capitalize()} Dataset Processing Results:")
    print(f"Successfully processed: {successful}")
    print(f"Failed: {failed}")
    print(f"Success rate: {(successful/(successful+failed)*100):.2f}%")
    
    # Print first few errors if any
    if error_log:
        print("\nFirst 5 errors encountered:")
        for error in error_log[:5]:
            print(error)
        
        # Save error log
        error_log_path = f"{dataset_type}_processing_errors.txt"
        with open(error_log_path, 'w') as f:
            f.write('\n'.join(error_log))
        print(f"\nComplete error log saved to: {error_log_path}")

def main():
    # Initialize YOLO model
    print("Loading YOLO model...")
    model = YOLO('yolov8n.pt')
    
    # Create output directories
    output_dirs = create_output_directories('processed_data')
    
    # Process Zebra Dataset
    print("\nProcessing Zebra Dataset...")
    process_dataset(
        zebra_dataset.df,
        'zebra',
        model,
        output_dirs['zebra'],
        base_path='data/StripeSpotter'
    )
    
    # Process Panda Dataset
    print("\nProcessing Panda Dataset...")
    # Test a single path first
    test_path = fix_panda_path(panda_dataset.df['path'].iloc[0], 'data')
    print(f"\nTesting panda path conversion:")
    print(f"Original path: {panda_dataset.df['path'].iloc[0]}")
    print(f"Converted path: {test_path}")
    print(f"Path exists: {os.path.exists(test_path)}")
    
    process_dataset(
        panda_dataset.df,
        'panda',
        model,
        output_dirs['panda'],
        base_path='data'
    )

if __name__ == "__main__":
    main()
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np
from tensorflow.keras.applications import VGG16
from tensorflow.keras.preprocessing import image
from typing import Tuple, Union
import os

# Force CPU usage
os.environ['CUDA_VISIBLE_DEVICES'] = '-1'

class SiameseNetwork:
    def __init__(self, input_shape: Tuple[int, int, int] = (224, 224, 3)):
        """
        Initialize the Siamese Network with VGG16 as the base model.
        
        Args:
            input_shape: Tuple of (height, width, channels) for input images
        """
        print("Initializing Siamese Network on CPU...")
        self.input_shape = input_shape
        self.model = self._create_siamese_network()
        
    def _create_siamese_network(self) -> models.Model:
        """
        Create the Siamese network architecture using VGG16 as the base model.
        
        Returns:
            Compiled Keras model
        """
        # Explicitly use CPU
        with tf.device('/CPU:0'):
            # Load pre-trained VGG16 model
            base_model = VGG16(weights='imagenet', 
                             include_top=False, 
                             input_shape=self.input_shape)
            
            base_model.trainable = False
            
            # Create siamese structure
            input1 = layers.Input(self.input_shape)
            input2 = layers.Input(self.input_shape)
            
            # Process both inputs through the same base model
            feature1 = base_model(input1)
            feature2 = base_model(input2)
            
            # Flatten features
            feature1 = layers.Flatten()(feature1)
            feature2 = layers.Flatten()(feature2)
            
            # Add dense layer for feature comparison
            dense_layer = layers.Dense(128, activation="relu")
            feature1 = dense_layer(feature1)
            feature2 = dense_layer(feature2)
            
            # Calculate absolute difference using Lambda layer
            distance = layers.Lambda(
                lambda tensors: tf.abs(tensors[0] - tensors[1])
            )([feature1, feature2])
            
            # Final prediction layer
            output = layers.Dense(1, activation='sigmoid')(distance)
            
            # Create and compile model
            model = models.Model(inputs=[input1, input2], outputs=output)
            
            # Use a smaller learning rate for stability
            optimizer = tf.keras.optimizers.Adam(learning_rate=0.0001)
            
            model.compile(
                optimizer=optimizer,
                loss='binary_crossentropy',
                metrics=['accuracy']
            )
            
            return model
    
    def train(self, 
              train_data: Tuple[Tuple[np.ndarray, np.ndarray], np.ndarray],
              validation_data: Tuple[Tuple[np.ndarray, np.ndarray], np.ndarray] = None,
              epochs: int = 10,
              batch_size: int = 32) -> None:
        """
        Train the Siamese network.
        
        Args:
            train_data: Tuple of ((x1, x2), y) where x1, x2 are input image pairs and y are labels
            validation_data: Optional validation data in same format as train_data
            epochs: Number of training epochs
            batch_size: Batch size for training
        """
        (x1, x2), y = train_data
        
        with tf.device('/CPU:0'):
            self.model.fit(
                [x1, x2], y,
                validation_data=validation_data,
                epochs=epochs,
                batch_size=batch_size
            )
    
    def predict_similarity(self, 
                         img1_path: str, 
                         img2_path: str) -> float:
        """
        Predict similarity between two images.
        
        Args:
            img1_path: Path to first image
            img2_path: Path to second image
            
        Returns:
            Similarity score between 0 and 1
        """
        # Validate image paths
        if not all(os.path.exists(path) for path in [img1_path, img2_path]):
            raise FileNotFoundError("One or both image paths are invalid")
            
        # Preprocess images
        img1 = self._preprocess_image(img1_path)
        img2 = self._preprocess_image(img2_path)
        
        # Predict similarity using CPU
        with tf.device('/CPU:0'):
            similarity = self.model.predict([img1, img2], verbose=0)
            return float(similarity[0][0])
    
    def _preprocess_image(self, 
                         image_path: str) -> np.ndarray:
        """
        Load and preprocess an image for the network.
        
        Args:
            image_path: Path to the image file
            
        Returns:
            Preprocessed image array
        """
        img = image.load_img(image_path, target_size=self.input_shape[:2])
        img_array = image.img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0)
        return img_array / 255.0
    
    def save_weights(self, path: str) -> None:
        """Save model weights to file."""
        self.model.save_weights(path)
    
    def load_weights(self, path: str) -> None:
        """Load model weights from file."""
        self.model.load_weights(path)

# Example usage
if __name__ == "__main__":
    try:
        # Initialize the network
        siamese_net = SiameseNetwork()
        print("Network initialized successfully!")
        
        # Example paths (replace with actual paths)
        img1_path = '/home/pinakinc/dev/umc-project/data/IPanda50/iPanda50-images/04_chengdui/04_135.jpg'
        img2_path = '/home/pinakinc/dev/umc-project/data/IPanda50/iPanda50-images/04_chengdui/04_210.jpg'
        
        if os.path.exists(img1_path) and os.path.exists(img2_path):
            similarity = siamese_net.predict_similarity(img1_path, img2_path)
            print(f"Similarity score: {similarity:.4f}")
        else:
            print("Please provide valid image paths")
            
    except Exception as e:
        print(f"Error: {str(e)}")
```
